```python
def distribute_crowd(row):
    total_crowd = row['Crowd size']
    event_type = row['Purpose of gathering']

    if event_type == 'Religious':
        entrance = int(total_crowd * np.random.uniform(0.05, 0.1))
        exit = int(total_crowd * np.random.uniform(0.05, 0.1))
        A = int(total_crowd * np.random.uniform(0.3, 0.4))
        B = int(total_crowd * np.random.uniform(0.3, 0.4))
        other = total_crowd - (entrance + exit + A + B)

    elif event_type == 'Sport':
        entrance = int(total_crowd * np.random.uniform(0.2, 0.25))
        exit = int(total_crowd * np.random.uniform(0.2, 0.25))
        A = int(total_crowd * np.random.uniform(0.2, 0.25))
        B = int(total_crowd * np.random.uniform(0.2, 0.25))
        other = total_crowd - (entrance + exit + A + B)

    elif event_type == 'Entertainment':
        entrance = int(total_crowd * np.random.uniform(0.25, 0.3))
        exit = int(total_crowd * np.random.uniform(0.25, 0.3))
        A = int(total_crowd * np.random.uniform(0.15, 0.2))
        B = int(total_crowd * np.random.uniform(0.15, 0.2))
        other = total_crowd - (entrance + exit + A + B)

    elif event_type == 'Political':
        entrance = int(total_crowd * np.random.uniform(0.15, 0.2))
        exit = int(total_crowd * np.random.uniform(0.15, 0.2))
        A = int(total_crowd * np.random.uniform(0.25, 0.3))
        B = int(total_crowd * np.random.uniform(0.25, 0.3))
        other = total_crowd - (entrance + exit + A + B)

    else:  # this will use for all the other things
        entrance = int(total_crowd * np.random.uniform(0.15, 0.2))
        exit = int(total_crowd * np.random.uniform(0.15, 0.2))
        A = int(total_crowd * np.random.uniform(0.25, 0.3))
        B = int(total_crowd * np.random.uniform(0.25, 0.3))
        other = total_crowd - (entrance + exit + A + B)

    return pd.Series([entrance, exit, A, B, other])
```

```python
df[['entrance', 'exit', 'A', 'B', 'other']] = df.apply(distribute_crowd, axis=1)
```
